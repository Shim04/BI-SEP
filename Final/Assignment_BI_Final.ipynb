```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('finance_liquor_sales.csv', parse_dates = ["date"])
```

```python
df.head()
```

```python
def get_lon(point):
    if pd.isna(point):
        return np.nan
    return re.findall('POINT \((.+) .+\)', point)[0]
```

```python
def get_lat(point):
    if pd.isna(point):
        return np.nan
    return re.findall('POINT \(.+ (.+)\)', point)[0]
```

```python
df['longitude'] = df['store_location'].apply(get_lon).astype(float)
df['latitude'] = df['store_location'].apply(get_lat).astype(float)
```

```python
df.to_csv('finance_liquor_sales_final.csv')
```
